```python
import pandas as pd

df=pd.read_csv("carp_2022.csv",encoding="shift_jis")
df
```

(1)先発登板数・救援登板数・勝利数が1以上の選手を, それぞれ多い順にソートした表を表示する.

```python
df_s1 = df[df["先発"]>=1].sort_values('先発', ascending=False)
df_s1
```

```python
df_s2 = df[df["救援"]>=1].sort_values('救援', ascending=False)
df_s2
```

```python
df_s3 = df[df["勝"]>=1].sort_values('勝', ascending=False)
df_s3
```

(2)被打率(被安打/被打数)・被本塁打率(被本塁打*9/投球回)を計算し, 背番号・選手名と併記した表を表示する.

```python
hit=df["被安打"]
atbat=df["対右被打数"]+df["対左被打数"]
home=df["被本塁打"]
inning=df["投球回"]

avg=hit/atbat
havg=home*9/inning

df2=df[["背番号","名前"]].copy()
df2["被打率"]=avg
df2["被本塁打率"]=havg
df2
```

(3)現役ドラフトで加入した戸根千明投手の, 今季の1軍成績を追加する.

```python
df.loc[27]=[49,"戸根千明",9,0,9,0,0,0,0,5.02,14.3333,8,8,63,16,4,9,3,0,1,27,10,3,3,27,6,6,0]
df
```

(4)対右・対左それぞれの被打率・K/BB(奪三振/与四球)を計算し, 被打率が小さい順にソートした表を表示する.

```python
from re import L
ratbat=df["対右被打数"]
rhit=df["対右被安打"]
rk=df["対右奪三振"]
rbb=df["対右与四球"]
latbat=df["対左被打数"]
lhit=df["対左被安打"]
lk=df["対左奪三振"]
lbb=df["対左与四球"]

ravg=rhit/ratbat
rkbb=rk/rbb
lavg=lhit/latbat
lkbb=lk/lbb

df3=df[["背番号","名前"]].copy()
df3["対右被打率"]=ravg
df3["対右K/BB"]=rkbb
df3["対左被打率"]=lavg
df3["対左K/BB"]=lkbb
```

```python
df3_s1 = df3.sort_values('対右被打率')
df3_s1
```

```python
df3_s2 = df3.sort_values('対左被打率')
df3_s2
```

一般的には左打者に対し左投手をワンポイントで登板させるといった采配が行われるが, 今季の村上選手については, 対右打率.305, 対左打率.346と, 相手投手の右左関係なく好成績を残している.
